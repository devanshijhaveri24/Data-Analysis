# Parsing the `categories` column

As noticed, the `categories` column is a dictionnary. We flatten this column and store it in a separate DataFrame.
This new DataFrame could be easily merged with the original one using the `datasetId` column.

```python
_dfs = []
for datasetId, category_row in df.set_index("datasetId")["categories"].iteritems():
    category_dict = eval(category_row) #evaluates each row dynamically
#column category has a dictionary of 2 keys: categories and type--- type values 'dataset', hence parsing categories
    categories = category_dict["categories"] #extracts key value categories
    _df = pd.DataFrame(categories) #create dataframe of categories
    _df["datasetId"] = datasetId #assign each datasetID
    _dfs.append(_df)
categories_df = pd.concat(_dfs, sort=False)
```

# Evolution of number of datasets per year and month

Exploring the evolution of the number of datasets over time. 
To make things more readable, I have resampled (using Pandas' [resample](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.resample.html) method) the dates to 
a frequency of one point per month. 

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 8))
(df.assign(date=lambda df: pd.to_datetime(df["dateUpdated"]))
                 .set_index('date')
                 .resample('1M')
                 .size()
                 .plot(ax=ax)
)

ax.set_ylabel("Number of datasets per month")

```

So far it seems that July (month number 7) is the one with the biggest number of datasets. 
This result might change once the August (and subsequent months) results are added.

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 8))
df['month']=pd.DatetimeIndex(df['dateUpdated']).month
data = df.groupby(df['month']).size().reset_index(name
    = 'number of datasets per month') 
sns.barplot(x='month', y='number of datasets per month', data=data)
plt.show()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 8))
df['month']=pd.DatetimeIndex(df['dateUpdated']).month
data = df.groupby(df['month']).size().reset_index(name
    = 'number of datasets per month') 
sns.lineplot(x='month', y='number of datasets per month', data=data)
plt.show()
```

# Cumulative number of datasets curve fitting
Curve fitting examines the relationship between one or more predictors (independent variables) and a response variable (dependent variable), with the goal of defining a "best fit" model of the relationship.

```python
from scipy.optimize import curve_fit
import numpy as np


def exp_f(x, a, b, c):
    return a*np.exp(b*x)+c

to_fit_s = (df.assign(date=lambda df: pd.to_datetime(df["dateUpdated"]))
                            .set_index('date')
                            .resample('1M')
                            .size()
                            .cumsum())

x = range(0, len(to_fit_s))
y = to_fit_s.values

fitted_params, covariance = curve_fit(exp_f, x, y, p0=(1, 1, 0))
fitted_s = pd.Series(exp_f(x, *fitted_params), index=to_fit_s.index)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 8))
fitted_s.plot(ax=ax, label='Fitted', style='o--')
to_fit_s.plot(ax=ax, label='Original', style='o-')
ax.legend()
ax.set_ylabel("Cumulative number of datasets")
```

# Merging the parsed **categories** column and the parsed **votes** column into one
The data needed to be pre-processed since the original categories column had null values which needed to be cleaned. Each sub-category needed to be mapped to its corresponding values (of the other columns).

```python
categories_df.shape
```

```python
df_filtered = categories_df
```

```python
occur = df_filtered.groupby(['datasetId']).size().reset_index(name='occurence')
display(occur)
type(occur)
```

```python
df1 = df[df['datasetId'].isin(occur['datasetId'])]
```

```python
df1.shape
```

```python
new_df = pd.merge(df1, occur, on='datasetId')
new_df.shape
```

```python
new_df.columns
```

```python
final_df = new_df.loc[new_df.index.repeat(new_df.occurence)]
```

```python
final_df.shape
```

```python
final_df.drop('occurence', axis=1)
```

```python
final_df.shape
```

```python
df_filtered = df_filtered.drop_duplicates(keep=False)
df_filtered.shape
```

```python
final_df = pd.merge(final_df, df_filtered, on='datasetId')
final_df.head()
```

```python
final_df = final_df.drop('categories', axis=1)
```

```python
final_df.head()
```

```python
final_df = final_df.drop_duplicates(keep='first', ignore_index=True)
```

```python
final_df.shape
```

```python
df_votes = final_df["voteButton"].apply(lambda x : dict(eval(x)) )
votes1 = df_votes.apply(pd.Series)
votes1.shape
```

```python
complete_df = pd.concat([final_df, votes1], axis=1).drop('voteButton', axis=1)
complete_df.head()
```
